# pca

```python
import os
import sys
from glob import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('~/Scripts/bin/v.mplstyle')

import pytraj as pt
```

```python

fname = 'step3-trajectories'
path = 'pca2'
```

```python
name = '4nt'

ds = pt.iterload(f'{path}/{name}/ds.nc', top=f'{path}/{name}/{fname}.parm7')
lds = len(ds) 

# ov = pt.iterload(f'{path}/ov.nc', top=f'{path}/{fname}.parm7')
# lov = len(ov)

traj = pt.iterload(f'{path}/{name}/{fname}.nc', top=f'{path}/{name}/{fname}.parm7')
data = pt.pca(traj, mask='@CA', n_vecs=10)

proj = data[0]
x = proj[0] 
y = proj[1] 

plt.scatter(x[:lds], y[:lds], marker='*', color='tab:blue', alpha=0.5, label='Splayed')    
plt.scatter(x[lds:], y[lds:], marker='o', color='tab:orange', alpha=0.5, label='Overhang')    

# Percent Variance 
pc1 = (data[1][0][0] / np.sum(data[1][0])) * 100 
pc2 = (data[1][0][1] / np.sum(data[1][0])) * 100 

plt.xlabel('PC1 (' + str(np.round(pc1, 1)) + ' %)')
plt.ylabel('PC2 (' + str(np.round(pc2, 1)) + ' %)')

lim=100
plt.xlim(-1*lim, lim)
plt.ylim(-1*lim, lim)
plt.legend()


plt.savefig(f'img/pca-{name}.png') # uncomment to save
```

```python
path = 'pca2/4nt'

ds = pt.iterload(f'{path}/{fname}.nc', top=f'{path}/{fname}.parm7')
data = pt.pca(ds, mask='@CA', n_vecs=10)
proj = data[0]
x = proj[0] 
y = proj[1] 
plt.scatter(x, y, marker='o', alpha=0.7, label='Splayed') 

ov = pt.iterload(f'{path}/ov.nc', top=f'{path}/{fname}.parm7')
data = pt.pca(ov, mask='@CA', n_vecs=10)
proj = data[0]
x = proj[0] 
y = proj[1] 
plt.scatter(x, y, marker='o', alpha=0.7,  label='Overhang') 

traj = pt.iterload(f'{path}/{fname}.nc', top=f'{path}/{fname}.parm7')
data = pt.pca(traj, mask='@CA', n_vecs=10)

proj = data[0]
x = proj[0] 
y = proj[1] 

plt.scatter(x, y, marker='*', c=range(traj.n_frames), alpha=0.15)    

# Percent Variance 
pc1 = (data[1][0][0] / np.sum(data[1][0])) * 100 
pc2 = (data[1][0][1] / np.sum(data[1][0])) * 100 

plt.xlabel('PC1 (' + str(np.round(pc1, 1)) + ' %)')
plt.ylabel('PC2 (' + str(np.round(pc2, 1)) + ' %)')

lim
plt.xlim(-1*lim, lim)
plt.ylim(-1*lim, lim)
cbar = plt.colorbar()
cbar.set_label('Frame #') 
plt.legend()


plt.savefig('img/pca-4ntcolor.png') # uncomment to save
```
